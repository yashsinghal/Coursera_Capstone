```python
import pandas as pd
import geocoder
from geopy.geocoders import Nominatim
import folium
import requests
import numpy as np
import matplotlib.cm as cm
import matplotlib.colors as colors
from sklearn.cluster import KMeans
```

## Get Data

### Get the list of sectors

```python
sectors = pd.read_html('https://www.myloancare.in/noida-circle-rate-revised/')[0]
sectors = sectors.drop('Builder Flats ( per sq meter )', axis=1)
sectors = sectors['Sectors'].str.replace('Sector', '')
sectors
sector_list = ''
for one_row in sectors:
    sector_list = sector_list + ", " + one_row
sector_list = sector_list.strip(', ')
sector_list = sector_list.replace(' ', '')
sector_list = sector_list.split(',')
sector_list
sector_name_list = ['Sector ' + str(one_sector.replace(',', '')) for one_sector in sector_list]
sector_name_list
```

### Get their Geocoder Coordinates

```python
# Get geocoder Coordinates
sector_lat = []
sector_lng = []
lat_lng_coords = None

locator = Nominatim(user_agent='myGeocoder')

for one_sector in sector_name_list:
    # loop until you get the coordinates
    sector_name = '{} Noida'.format(one_sector)
    print(sector_name)
    location = locator.geocode(sector_name)
    

    sector_lat.append(location.latitude)
    sector_lng.append(location.longitude)
```


Store this in a Dataframe


```python
sector_df = pd.DataFrame(list(zip(sector_name_list, sector_lat, sector_lng)),  columns=['Sector', 'Latitude', 'Longitude'])
sector_df['Sector'] = sector_df['Sector'].str.capitalize()
sector_df
```

### Get Noida Coordinates

```python
noida_loc = locator.geocode('Noida')
latitude = noida_loc.latitude
longitude = noida_loc.longitude
print(latitude)
print(longitude)
```

## Map the Points

```python
noida_map = folium.Map(location=[latitude, longitude], zoom_start=13)

# add markers to map
for lat, lng, sector in zip(sector_df['Latitude'], sector_df['Longitude'], sector_df['Sector']):
    label = sector
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#666666',
        fill_opacity=0.7,
        parse_html=False).add_to(noida_map)  
    
noida_map
```

```python
# Function to get nearby venues
def getNearbyVenues(names, latitudes, longitudes, radius):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Sector', 
                  'Sector Latitude', 
                  'Sector Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

### One Hot Encode

We one hot encode the data for further processing and clustering

```python
# one hot encoding
noida_onehot = pd.get_dummies(noida_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
noida_onehot['Sector'] = noida_venues['Sector'] 

# move neighborhood column to the first column
fixed_columns = [noida_onehot.columns[-1]] + list(noida_onehot.columns[:-1])
noida_onehot = noida_onehot[fixed_columns]

noida_onehot
```

### Group Data

We group the data for each sector so we can cluster the sectors

```python
noida_grouped = noida_onehot.groupby('Sector').mean().reset_index()
noida_grouped_drop = noida_grouped.drop('Sector', 1)

# add a column that counts the number of unique categories present in each sector
noida_grouped_count = noida_grouped_drop[noida_grouped_drop>0].count(axis=1, numeric_only=True)/25
noida_grouped_list = noida_grouped_count.values.tolist()

noida_grouped.insert(loc=1, column='Category Count', value=noida_grouped_list)
noida_grouped
```

### Clustering
We divide the data into 3 clusters which are expected to be:
1. Fully Self Sufficient - Have low or none dependence on other sectors
2. Partly Self Sufficient - Have partial dependence on other sectors
3. Barely Self Sufficient - Have high dependence on other sectors

```python
# set number of clusters
kclusters = 3

noida_grouped_clustering = noida_grouped.drop('Sector', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=42).fit(noida_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:89]
```

We add the cluster labels to the grouped sectors data and merge it with sector_df to get the lat and long of each sector

```python
# add clustering labels


noida_grouped_sectors = noida_grouped
noida_grouped_sectors.insert(1, 'Cluster Labels', kmeans.labels_)
noida_grouped_sectors

# # merge noida_grouped_sectors with sector_df to add latitude/longitude for each neighborhood
noida_grouped_sectors = noida_grouped_sectors.join(sector_df.set_index('Sector'), on='Sector')

noida_grouped_sectors.head() # check the last columns!
```

### Mapping Clusters

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=12)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(noida_grouped_sectors['Latitude'], noida_grouped_sectors['Longitude'], noida_grouped_sectors['Sector'], noida_grouped_sectors['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
